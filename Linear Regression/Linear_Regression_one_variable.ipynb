```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Population = np.array([61101,55277,85186,70032,58598,83829,74764,85781,64862,50546,57107,141640,57340,84084,56407,53794,63654,51301,64296,70708])
```

```python
Profit = np.array([175920,91302,136620,118540,68233,118860,43483,120000,65987,38166,32522,155050,31551,72258,7161,35129,53048,5607,36518,53893])
```

```python
%matplotlib inline
plt.xlabel('Population.')
plt.ylabel('Price')
plt.scatter(Population,Profit,color='red',marker='+')
```

```python
linreg = LinearRegression()
```

```python
Population = Population.reshape(-1,1)
```

```python
linreg.fit(Population,Profit)
```

```python
P_pred = linreg.predict(Population)
```

```python
plt.scatter(Population,Profit)
plt.plot(Population,P_pred,color="blue")
plt.show()
```

```python
print(linreg.coef_)
```

```python
print(linreg.intercept_)
```
